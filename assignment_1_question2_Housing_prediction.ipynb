```python
def cost_calculator(x,y,w,b):

    cost=0

    for i in range(x.shape[0]):

        f_wb=np.dot(x.iloc[i],w) + b

#         cost calculator
        cost+=(f_wb-y.iloc[i])**2

#     dividing by number of events, as it is in mean squared error formula
    cost=cost/x.shape[0]

    return cost
```

```python
cost_calculator(X_train, y_train, w, b)
```

```python
def gradient_calculator(x,y,w,b):

    dw=0
    db=0

    for i in range(x.shape[0]):

        f_wb=np.dot(x.iloc[i],w)+b

#       differential of cost w.r.t w and we will divide the sum later on by x.shape[0]
        dw_i=sum((np.dot(X_train.iloc[i],w)+b-y_train.iloc[i])*X_train.iloc[i])
#       sum of differentials
        dw+=dw_i
#     differential of cost w.r.t bias and we will divide the sum later on by x.shape[0]
        db_i=(f_wb-y.iloc[i])
#     sum of bias differential
        db+=db_i
#     division by x.shape[0]
    dw=dw/x.shape[0]
    db=db/x.shape[0]

    return dw, db
```

### updating weights and bias using batch gradient descent

```python
def gradient_updation(x,y,w,b, cost_function=cost_calculator,
                      gradient_function=gradient_calculator,  iterations=100000, alpha=0.1):

    cost_history=[]
    w_history=[]

    for i in range(iterations):
        dw, db=gradient_function(x,y,w,b)
#       updating the parameters
        w=w-alpha*dw

        b=b-alpha*db

#       saving the cost for each iteration
        if i<100000:
            cost=cost_function(x,y,w,b)
            cost_history.append(cost)
        # Print cost every at intervals 20 times or as many iterations if < 20
        if i% math.ceil(iterations/100)==0:
            w_history.append(w)
            print(f"Iteration {i:4}: Cost {float(cost_history[-1]):8.5f}   ")

    return w,b, cost_history, w_history
```
